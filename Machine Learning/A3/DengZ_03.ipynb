```python
def combine_stock_code_country(df, country):
    """

    return a new dataframe that in the same country, each invoice has combined, has some new stock code
    """
    country_df = df[df.Country == country]


    invoicno_country = country_df.InvoiceNo.value_counts()




    new_df = pd.DataFrame([], columns = ['new_invoice_no', 'new_stock_code'])

    new_invoice = invoicno_country.index
    new_df['new_invoice_no'] = new_invoice
    new_stock_code = []
    for d in invoicno_country.index: 

        stockcode = []

        for i in (country_df[country_df.InvoiceNo == d]).StockCode:
            stockcode.append(str(i))
        new_stock_code.append(stockcode)
        # print(stockcode)
        

    new_df['new_stock_code'] = new_stock_code


    return new_df, new_stock_code
```
