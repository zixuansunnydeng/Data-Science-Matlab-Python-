##2.2 Soft [5 points]
Now train a soft-margin SVM classifier for the same problem. Use crossvalidation to select the best value of C. Did you achieve better results?

**run time around 3 min**

```python
# crete kfold object with 5 splits
folds = KFold(n_splits = 5, shuffle = True, random_state = 1)
hyper_params = [{'C':[1,5,10]}]
model = svm.SVC()
# set GridSearchCV()
model_cv = GridSearchCV(estimator = model, 
                        param_grid = hyper_params, 
                        scoring= 'accuracy', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)      

# fit the model
model_cv.fit(X_train_new, y_train_new)
```

```python
# cv results
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
# get best hyperparameter 
best_accuracy = model_cv.best_score_
best_hyperparams = model_cv.best_params_
print(best_accuracy, best_hyperparams)
```

```python
#best tuned model
best_model_soft = svm.SVC(C=5)
best_model_soft.fit(X_train_new,y_train_new)
y_predict_soft = best_model_soft.predict(X_test_new)
print("accuracy", accuracy_score(y_test_new, y_predict_soft))
```

### Radial Basis Function Kernel
