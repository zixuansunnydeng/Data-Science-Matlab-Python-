## Uber-SVM [20 points]

Sometimes it is worth using different slack variables for different classes in the soft-margin SVM
formulation. For example, we will use this trick if the datasets are not balanced, or if it is more
important to correctly classify one class instead of the other. More formally, for a binary classification
problem between two classes + and âˆ’, we would like to optimize the following function:
 

$\min\limits_{\lambda,b} \frac{1}{2}\|{\omega}\|^{2}  + C_+\sum\limits_{i:y_i is +}\xi_i + C_-\sum\limits_{i:y_i is -}\xi_i   $
\
such that $y_i(\omega^Tx_i+b) \geq 1-\xi_i$ and $\xi_i \geq 0   $ for any $i$. Your job is to derive the Lagrangian dual formulation of this problem.

My solution


$\min\limits_{\lambda,b} \frac{1}{2}\|{\omega}\|^{2}  + C_+\sum\limits_{i:y_i is +}\xi_i + C_-\sum\limits_{i:y_i is -}\xi_i   $

We can rewrite the constrains condition as 
$ y_i(\omega^Tx_i+b) \geq 1-\xi_i \Rightarrow y_i(\omega^Tx_i+b) - 1 + \xi_i  \geq 0$



Lagrangian function: $\mathscr{L}(\omega, b, \xi, \lambda) = \frac{1}{2}\|{\omega}\|^{2}  + C_+\sum\limits_{i:y_i is +}\xi_i + C_-\sum\limits_{i:y_i is -}\xi_i - \sum\limits_{i}\lambda_i[y_i(\omega^Tx_i+b) - 1 + \xi_i] $

Now, the optimal objective is:

$$ \min\limits_{\omega,b} \max\limits_{\lambda} \mathscr{L}(\omega, b, \xi, \lambda) \quad s.t. \quad \lambda_i \geq 0,\quad \forall i \leq  n 
$$




$$ \frac{\partial \mathscr{L}(\omega, b, \xi, \lambda )}{\partial \omega} = \omega - \sum_{i}\lambda_iy_ix_i; \frac{\partial \mathscr{L}(\omega, b, \xi, \lambda )}{\partial \omega} = 0 \Rightarrow \omega = \sum_{i}\lambda_iy_ix_i $$

$$  $$


$$ \frac{\partial \mathscr{L}(\omega, b, \xi, \lambda )}{\partial b} = \sum_{i}\lambda_iy_i;  \frac{\partial \mathscr{L}(\omega, b, \xi, \lambda )}{\partial b} = 0 \Rightarrow \sum_{i}\lambda_iy_i = 0$$


$$ \frac{\partial \mathscr{L}(\omega, b, \xi, \lambda )}{\partial \xi} = C_+ - \lambda_i + C_- -\lambda_i ;  \frac{\partial \mathscr{L}(\omega, b, \xi, \lambda )}{\partial \xi} = 0 \Rightarrow C_+ - \lambda_i + C_- -\lambda_i = 0
$$
\
Thus we get
$$ \lambda_i = 
\left\{
  \begin{array}{@{}ll@{}}
    C_+, & \text{if}\ i: y_i \quad \text{is} +\\
    C_-, & \text{otherwise}
  \end{array}\right.
$$

Plugging these back to the Lagrangian function, rearranging terms we obtain the dual:

$$\max\limits_{\lambda} \sum_{i} \lambda_i - \frac{1}{2} \sum_{i,j} y_i y_j \lambda_i \lambda_j(x_i^Tx_j)     $$
s.t. 
$$\lambda_i = 
\left\{
  \begin{array}{@{}ll@{}}
    C_+, & \text{if}\ i: y_i \quad \text{is} +\\
    C_-, & \text{otherwise}
  \end{array}\right. and \sum_{i} \lambda_i y_i = 0$$
