```python
def prior_probability_class(df, category):
    yes_category = sum(df.iloc[:,0]==category)
    no_category = sum(df.iloc[:,0]!=category)
    prob_category = yes_category / (yes_category+no_category)
    return prob_category

print(prior_probability_class(df, 'spam'))
```

```python
def naive_baye(training_df, testing_df):
    predict_label = []


    # msg = testing_df.iloc[:,1]

    initial_condition_p_spam = prior_probability_class(training_df, 'spam')
    initial_condition_p_ham = prior_probability_class(training_df, 'ham')

    conditional_p_spam = []
    conditional_p_ham = []

    # testing_df_spam = testing_df[testing_df.iloc[:,0] == 'spam']
    # testing_df_ham = testing_df[testing_df.iloc[:,0] == 'ham']


    p_word_spam = conditional_probability(df, 'spam')
    p_word_ham = conditional_probability(df, 'ham')

    for row_index in range(testing_df.shape[0]):
        msg = testing_df.iloc[row_index,1]
        msg_list = msg.split(' ') # ['go', 'bye', 'machine' 'learning']
        for word in msg_list:  # 'go'
            # print(word)
            if word in p_word_spam:
                  initial_condition_p_spam = initial_condition_p_spam * p_word_spam[word]
    
        conditional_p_spam.append(initial_condition_p_spam)


    for row_index in range(testing_df.shape[0]):
        msg = testing_df.iloc[row_index,1]
        msg_list = msg.split(' ') # ['go', 'bye', 'machine' 'learning']
        for word in msg_list:  # 'go'
            if word in p_word_ham:
                initial_condition_p_ham = initial_condition_p_ham * p_word_ham[word]
    
        conditional_p_ham.append(initial_condition_p_ham)

    # p_spam = prior_probability_class(training_df, 'spam')
    # p_ham = prior_probability_class(training_df, 'ham')
    # if p_msg_spam * p_spam > p_msg_ham * p_ham: 
    # print(len(conditional_p_ham))
    # print(len(conditional_p_spam))

    for i in range(len(conditional_p_ham)):

        if conditional_p_spam[i] > conditional_p_ham[i]:
            predict_label.append('spam')
        else: 
            predict_label.append('ham')
        
    
    return predict_label


def accuracy(true_label, predict_label):
    total_num = true_label.shape[0]
    result_matrix = true_label == predict_label 
    correct = sum(result_matrix)
    accuracy = correct / total_num 

    return accuracy
```
