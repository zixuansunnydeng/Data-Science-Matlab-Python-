```python
# li = []
# for column in X.columns: 
#     li.append(X[column].dtype != 'object')
# print(li)
# new_X = X.iloc[:,li]


columns_demmy = ['Education','Religion','EthnicMale','EthnicFemale']
for i in columns_demmy:
    df = df.join(pd.get_dummies(df[i]))
    # print(pd.get_dummies(df[i]))
df = df.drop(['Education', 'Religion','EthnicMale','EthnicFemale','State', 'County'], axis=1)
df.head()

X = df.loc[:, df.columns != 'Democrat']
y = df['Democrat']
```

```python
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)

rforest = RandomForestClassifier(n_estimators=50, max_depth=3, random_state=0)
rforest.fit(X_train,y_train)
y_predict = rforest.predict(X_test)
print("the accuracy is",metrics.accuracy_score(y_test, y_predict))   
```

Random forests better, around 0.1%. When dataset are large, random forest will perform well. A random forest is a collection of decison tree, it has the most vote by majority. And therefore, it is usually have better accuracy than decison tree.
