We need to convert categorical data to one-hot encoding. There are four features that are categorical: Education, Religion, EthnicMale, EthnicFemale. 

```python
columns_demmy = ['Education','Religion','EthnicMale','EthnicFemale']
for i in columns_demmy:
    df = df.join(pd.get_dummies(df[i]))
    # print(pd.get_dummies(df[i]))

df.head()

```

```python
# drop the categorical feature
# df = df.drop(['Education', 'Religion','EthnicMale','EthnicFemale'], axis=1)
# df.head()
df.drop('State', inplace=True, axis=1)
df.drop('County', inplace=True, axis=1)
```

```python
# implement the decison tree with scikit-learn package. 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix

X = df.iloc[:,20:34]
y = df['Democrat']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)
tree_clf = DecisionTreeClassifier(max_depth = 3, max_leaf_nodes=50, random_state=0)
tree_clf.fit(X_train,y_train)
y_predict = tree_clf.predict(X_test)
print("the accuracy is",metrics.accuracy_score(y_test, y_predict))   
```
